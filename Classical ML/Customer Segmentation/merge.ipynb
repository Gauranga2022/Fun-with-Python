```python
import nbformat
import os

# Define the path to your notebooks folder in Google Drive.
notebooks_folder = '/content/drive/MyDrive/Customer Segmentation/'

# List the notebook filenames in the order you want to merge them.
notebook_files = [
    '1. Customer Segmentation - (Data Science).ipynb',
    '2. Customer Segmentation - (Model Building-K-Means Clustering).ipynb',
    '3. Customer Segmentation - (Model Building-Agglomerative Clustering).ipynb',
    '4. Customer Segmentation - (Model Building-Gaussian Mixture Clustering).ipynb',
    '5. Customer Segmentation - (Model Building-Mini Batch K-Means Clustering).ipynb',
    '6. Customer Segmentation - (Model Building-DBSCAN Clustering).ipynb',
    '7. Customer Segmentation - (Conclusion).ipynb'
]

# Create a new empty notebook for the merged output.
merged_notebook = nbformat.v4.new_notebook()

# Loop through and merge the notebooks.
for nb_name in notebook_files:
    nb_path = os.path.join(notebooks_folder, nb_name)
    with open(nb_path, 'r', encoding='utf-8') as f:
        nb = nbformat.read(f, as_version=4)
    # Append the cells from the current notebook into the merged notebook.
    merged_notebook.cells.extend(nb.cells)

# Define the output path for the merged notebook.
output_file = os.path.join(notebooks_folder, 'Customer_Segmentation.ipynb')
with open(output_file, 'w', encoding='utf-8') as f:
    nbformat.write(merged_notebook, f)

print("Merged notebook saved to:", output_file)
```
